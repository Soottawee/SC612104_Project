### ข้อ 1: OOP — class BankAccount
สร้าง class `BankAccount` ที่มี:
1. `__init__(self, owner, balance=0)` — attribute `owner` และ `balance`
2. method `deposit(self, amount)` — เพิ่มเงินเข้า balance
3. method `withdraw(self, amount)` — ลดเงินออกจาก balance (ถ้า amount มากกว่า balance ให้ print ว่า "เงินไม่พอ" และไม่หักเงิน)
4. method `show_balance(self)` — แสดงยอดเงินปัจจุบัน

ทดสอบสร้าง object แล้วลองเรียก deposit/withdraw หลายครั้ง


````python
# เขียนคำตอบที่นี่

class BankAccount:
    def __init__(self, owner, balance=0):
        self.owner = owner
        self.balance = balance

    def deposit(self, amount):
        if amount > 0:
            self.balance += amount
            print(f"ฝากเงิน {amount} บาท. ยอดเงินคงเหลือ: {self.balance} บาท")
        else:
            print("จำนวนเงินที่ฝากต้องมากกว่า 0")

    def withdraw(self, amount):
        if amount > 0:
            if self.balance >= amount:
                self.balance -= amount
                print(f"ถอนเงิน {amount} บาท. ยอดเงินคงเหลือ: {self.balance} บาท")
            else:
                print(f"เงินไม่พอ! ยอดเงินปัจจุบัน: {self.balance} บาท แต่ต้องการถอน {amount} บาท")
        else:
            print("จำนวนเงินที่ถอนต้องมากกว่า 0")

    def show_balance(self):
        print(f"บัญชีของคุณ {self.owner} มียอดเงินคงเหลือ: {self.balance} บาท")


# ทดสอบการใช้งาน BankAccount class
print("--- ทดสอบ BankAccount ---")
my_account = BankAccount("สมใจ", 1000)
my_account.show_balance()

my_account.deposit(500)
my_account.show_balance()

my_account.withdraw(200)
my_account.show_balance()

my_account.withdraw(1500) # ทดสอบกรณีเงินไม่พอ
my_account.show_balance()

my_account.deposit(0) # ทดสอบกรณีฝาก 0 บาท
my_account.withdraw(0) # ทดสอบกรณีถอน 0 บาท
````

### ข้อ 2: OOP — เชื่อมกับ list ของ object
จาก class `BankAccount` ในข้อ 3 ให้สร้าง object 3 บัญชี เก็บไว้ใน list จากนั้น:
1. วน loop deposit เงิน 100 บาทให้ทุกบัญชี
2. ใช้ list comprehension หาบัญชีที่มี balance มากกว่า 500 บาท
3. หาบัญชีที่มี balance สูงสุด (Hint: ใช้ `max()` กับ parameter `key`)


````python
# เขียนคำตอบที่นี่

bank_accounts = [
    BankAccount("บัญชีA", 200),
    BankAccount("บัญชีB", 450),
    BankAccount("บัญชีC", 600)
]

print("--- ยอดเงินเริ่มต้น ---")
for account in bank_accounts:
    account.show_balance()

print("\n--- ฝากเงิน 100 บาททุกบัญชี ---")
for account in bank_accounts:
    account.deposit(100)

print("\n--- ยอดเงินหลังฝาก ---")
for account in bank_accounts:
    account.show_balance()

accounts_over_500 = [account for account in bank_accounts if account.balance > 500]
print("\n--- บัญชีที่มีเงินมากกว่า 500 บาท ---")
if accounts_over_500:
    for account in accounts_over_500:
        account.show_balance()
else:
    print("ไม่พบบัญชีที่มีเงินมากกว่า 500 บาท")

top_account = max(bank_accounts, key=lambda acc: acc.balance)
print("\n--- บัญชีที่มีเงินสูงสุด ---")
top_account.show_balance()
````

### ข้อ 3: zoo
1. สร้าง Class สัตว์ต่อไปนี้ ได้แก่  `Dog`, `Cat`, `Lion` ซึ่งมีชื่อ `name`, สี `color`, ขนาดตัว `size` (หนึ่งใน `large`, `medium`, `small`)
2. ในแต่ละ Class เมื่อเรียกฟังก์ชัน `makeSound()` แล้ว แต่ละคลาสจะร้องเสียงสัตว์นั้น ๆ
3. สร้าง list ของ `animals` หลายประเภท ที่เมื่อ loop เข้าไปทีละตัว แล้วจะบอกชื่อ พร้อมเสียงร้อง (ใช้ `__str__`)
4. ใน list ของ `animals` ที่สร้างไว้ ให้วนเพื่อหาว่า สัตว์ทั้งหมด มีกี่ตัวที่มีขนาดต่าง ๆ

````python
class Animal:
    def __init__(self, name, color, size):
        self.name = name
        self.color = color
        self.size = size

    def makeSound(self):
        raise NotImplementedError("คลาสย่อยต้องมีการ implements เมธอดนี้")

    def __str__(self):
        try:
            sound = self.makeSound()
        except NotImplementedError:
            sound = "(ไม่ได้กำหนดเสียง)"
        return f"{self.name} (สี: {self.color}, ขนาด: {self.size}) ร้องว่า {sound}"

class Dog(Animal):
    def makeSound(self):
        return "โฮ่ง"

class Cat(Animal):
    def makeSound(self):
        return "เหมียว"

class Lion(Animal):
    def makeSound(self):
        return "โฮกปิ๊บ"

animals = [
    Dog("ตูบ", "น้ำตาล", "กลางๆ"),
    Cat("นวล", "ขาว", "จิ๋วหลิว"),
    Lion("ซิมบ้า", "ทอง", "ใหญ่"),
    Dog("โอโม่", "ดำ", "ใหญ่"),
    Cat("เหมียว", "สามสี", "จิ๋วหลิว"),
    Lion("อาคุมามาทาท่า", "ทอง", "กลางๆ"),
    Dog("ด่าง", "เทา", "กลางๆ")
]

print("--- รายชื่อสัตว์และเสียงร้อง ---")
for animal in animals:
    print(animal)

print("\n--- การนับจำนวนสัตว์ตามขนาด ---")
size_counts = {"ใหญ่": 0, "กลางๆ": 0, "จิ๋วหลิว": 0}

for animal in animals:
    if animal.size in size_counts:
        size_counts[animal.size] += 1
    else:
        print(f"คำเตือน: ไม่พบขนาด '{animal.size}' สำหรับ {animal.name}")

for size, count in size_counts.items():
    print(f"จำนวนสัตว์ขนาด {size}: {count} ตัว")
````
