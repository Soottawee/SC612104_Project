### ข้อ 2: Operators
กำหนด `a = 23` และ `b = 4` จงคำนวณและ print ผลลัพธ์ต่อไปนี้:
- ผลรวมของ a และ b
- ผลหารแบบปัดเศษทิ้งของ a หาร b
- เศษที่เหลือจากการหาร a ด้วย b
- a ยกกำลัง 2


````python
# เขียนคำตอบข้อ 2 ที่นี่
a = 23
b = 4

print(f"a + b = {a + b}")
print(f"a // b = {a // b}")
print(f"a % b = {a % b}")
print(f"a ** 2 = {a ** 2}")
````

### ข้อ 4: if/elif/else — ตรวจ BMI
สูตรคำนวณ BMI คือ `น้ำหนัก(kg) / (ส่วนสูง(m) ** 2)`

จงเขียนโค้ดที่:
1. กำหนดตัวแปร `weight = 65` (kg) และ `height = 1.70` (m)
2. คำนวณค่า BMI
3. ใช้ if/elif/else แปลผลตามเกณฑ์นี้:
   - BMI < 18.5 → "น้ำหนักน้อย"
   - 18.5 ≤ BMI < 23 → "น้ำหนักปกติ"
   - 23 ≤ BMI < 25 → "น้ำหนักเกิน"
   - BMI ≥ 25 → "อ้วน"
4. print ค่า BMI (ทศนิยม 2 ตำแหน่ง) และผลแปลความหมาย

💡 Hint: ใช้ `round(ค่า, 2)` เพื่อปัดทศนิยม 2 ตำแหน่ง


````python
# เขียนคำตอบข้อ 4 ที่นี่
weight = 65
height = 1.70

bmi = weight / (height ** 2)
BMI = round(bmi, 2)

if bmi < 18.5:
  print(f"{BMI} น้ำหนักน้อย")

elif bmi <= 23:
    print(f"{BMI} น้ำหนักปกติ")

elif 23 <= bmi < 25:
    print(f"{BMI} น้ำหนักเกิน")

elif bmi >= 25:
    print(f"{BMI} จ้ำม่ำ")
````

### ข้อ 5: ท้าทายเพิ่มเติม (Challenge)
เขียนโปรแกรมเช็คว่าปี (year) ที่กำหนดเป็น**ปีอธิกสุรทิน** (leap year) หรือไม่ โดยใช้เงื่อนไข:
- หารด้วย 4 ลงตัว **และ** (หารด้วย 100 ไม่ลงตัว **หรือ** หารด้วย 400 ลงตัว) → เป็นปีอธิกสุรทิน

ทดสอบกับปี 2024, 2023, 2000, 1900


````python
# เขียนคำตอบข้อ 5 ที่นี่
year = 1900

if year % 400 == 0:
  print(f"เป็นปีอธิกสุรทิน")

elif year % 100 == 0:
    print(f"ไม่เป็นปีอธิกสุรทิน")

elif year % 4 == 0:
    print(f"เป็นปีอธิกสุรทิน")

else:
    print(f"ไม่เป็นปีอธิกสุรทิน")
````
