### ข้อ: tuple — unpacking
มีข้อมูลพิกัด GPS เก็บเป็น list ของ tuple: `locations = [(13.7, 100.5), (18.8, 98.9), (7.9, 98.3)]`

จงเขียน `for` loop ที่ unpack แต่ละ tuple เป็น `lat, lon` แล้ว print ในรูปแบบ `"ละติจูด: ..., ลองจิจูด: ..."`


````python
# เขียนคำตอบที่นี่
locations = [(13.7, 100.5), (18.8, 98.9), (7.9, 98.3)]
for lat, lon in locations:
    print(f"ละติจูด: {lat}, ลองจิจูด: {lon}")
````

### ข้อ: dict — ระบบเก็บคะแนน
สร้าง dict เก็บคะแนนสอบของนักศึกษา 4 คน (ตั้งชื่อและคะแนนเองได้)

1. วน loop print ชื่อและคะแนนทุกคนในรูปแบบ `"ชื่อ: คะแนน"`
2. หาว่าใครได้คะแนนสูงสุด (Hint: ใช้ `.items()` ร่วมกับ `if` ใน loop เพื่อเก็บค่าสูงสุดที่เจอ)
3. เพิ่มนักศึกษาใหม่อีก 1 คนเข้าไปใน dict


````python
# เขียนคำตอบที่นี่
scores = {("Mha", 79), ("Mhee", 58), ("Mhoo", 87), ("Mhor", 98)}
for name, score in scores:
    print(f"ชื่อ: {name} ได้ {score} คะแนน")
````

### ข้อ: Comprehension
จาก list คะแนนสอบ `scores = [45, 67, 82, 91, 58, 73, 39, 95, 60]`

1. ใช้ **list comprehension** สร้าง list ใหม่ที่มีแต่คะแนนที่ผ่าน (>= 50)
2. ใช้ **list comprehension** แปลงคะแนนทั้งหมดเป็นเกรด (>= 80 = "A", >= 70 = "B", >= 60 = "C", >= 50 = "D", น้อยกว่านั้น = "F") — Hint: ใช้ conditional expression `x if condition else y` ข้างใน comprehension
3. ใช้ **dict comprehension** สร้าง dict จับคู่ index กับคะแนน เช่น `{0: 45, 1: 67, ...}` (Hint: ใช้ `enumerate()`)


````python
# เขียนคำตอบที่นี่
scores = [45, 67, 82, 91, 58, 73, 39, 95, 60]
scores_pass = [score for score in scores if score >= 50]
print(scores_pass)
print()
grades = [
    "A" if score >= 80 else
    "B" if score >= 70 else
    "C" if score >= 60 else
    "D" if score >= 50 else
    "F"
    for score in scores]
for score, grade in zip(scores, grades):
    print(f"{score} คะแนน เกรด {grade}")
scores_dict = {index: score for index, score in enumerate(scores)}
print(scores_dict)
````

### ข้อ: ท้าทายเพิ่มเติม (Challenge) — รวมทุกอย่างเข้าด้วยกัน
มีข้อมูลนักศึกษาเก็บแบบนี้:

```python
names = ["มานี", "สมชาย", "วิชัย", "สมหญิง"]
midterm_scores = [78, 85, 62, 90]
final_scores = [82, 80, 70, 95]
```

จงเขียนโค้ดที่:
1. ใช้ `zip()` รวม 3 list เข้าด้วยกัน คำนวณคะแนนเฉลี่ย (midterm + final) / 2 ของแต่ละคน
2. สร้าง dict ผลลัพธ์ในรูปแบบ `{"มานี": 80.0, "สมชาย": 82.5, ...}` ด้วย **dict comprehension**
3. ใช้ `.items()` วน loop หาว่าใครได้คะแนนเฉลี่ยสูงสุด แล้ว print ชื่อกับคะแนนนั้นออกมา


````python
# เขียนคำตอบที่นี่
names = ["มานี", "สมชาย", "วิชัย", "สมหญิง"]
midterm_scores = [78, 85, 62, 90]
final_scores = [82, 80, 70, 95]

averages=[]
for name,mid,final in zip(name,midterm_scores,final_scores):
  averages.append((mid+final)/2)

avg_dict={
    name:(mid+final)/2
    for name,mid,final in zip(names,midterm_scores,final_scores)
}

max_name= ""
max_score=0
for name,score in avg_dict.items():
  if score > max_score:
    max_score=score
    max_name=name
print(f"นักเรียนได้คะแนนเฉลี่ยสูงสุด คือ {max_name} ได้ {max_score}")
````

### ข้อ: range() และ for loop พื้นฐาน
ใช้ `for` loop กับ `range()` เพื่อ print เลขคู่ทั้งหมดตั้งแต่ 0 ถึง 20 (รวม 20 ด้วย)

💡 Hint: ใช้ `range()` แบบ 3 พารามิเตอร์ (start, stop, step)

````python
# เขียนคำตอบที่นี่
for i in range(0,21,2):
  print(i)
````

### ข้อ: Accumulator Pattern
เขียน `for` loop หาผลรวมของเลขคูณ 3 ทั้งหมดที่อยู่ระหว่าง 1 ถึง 50 (เช่น 3, 6, 9, ..., 48)

💡 Hint: ใช้ `%` เพื่อเช็คว่าหารด้วย 3 ลงตัวไหม


````python
# เขียนคำตอบที่นี่
total = 0
for i in range(1,51):
  if i%3==0:
    total+=i
print(total)
````

### ข้อ: while loop
เขียน `while` loop ที่เริ่มจากเลข 100 แล้วหารด้วย 2 ไปเรื่อยๆ (ปัดเศษทิ้งด้วย `//`) จนกว่าค่าจะน้อยกว่า 1 — print ค่าทุกรอบ

ตัวอย่างผลลัพธ์ที่ควรได้: `100, 50, 25, 12, 6, 3, 1`


````python
# เขียนคำตอบที่นี่
number = 100
while number >=1:
  print(number)
  number//=2
````

### ข้อ: break และ continue
มีรายชื่อสินค้าในสต็อก ต้องการหาสินค้าตัวแรกที่ "หมดสต็อก" (จำนวน = 0) แล้วหยุดค้นหาทันทีที่เจอ — ถ้าสินค้าไม่ได้หมด (จำนวน > 0) ให้ print ชื่อและจำนวนสินค้านั้นออกมาก่อน

```python
stock = {"ปากกา": 15, "สมุด": 8, "ยางลบ": 0, "ไม้บรรทัด": 5}
```

💡 Hint: ใช้ `.items()` เพื่อวน loop ทั้ง key และ value ของ dict พร้อมกัน เช่น `for name, qty in stock.items():`


````python
# เขียนคำตอบที่นี่
stock = {"ปากกา": 15, "สมุด": 8, "ยางลบ": 0, "ไม้บรรทัด": 5}
for name, qty in stock.items():
  if qty ==0:
    print(f"พบสินค้าหมดสต็อก:{name}")
    break
  print(name,qty)
````

### ข้อ: ท้าทายเพิ่มเติม (Challenge) — FizzBuzz
เขียนโปรแกรมวนเลข 1 ถึง 30 โดยมีเงื่อนไข:
- ถ้าหารด้วย 3 ลงตัว → print "Fizz"
- ถ้าหารด้วย 5 ลงตัว → print "Buzz"
- ถ้าหารด้วยทั้ง 3 และ 5 ลงตัว → print "FizzBuzz"
- นอกจากนั้น → print ตัวเลขนั้นตามปกติ

(โจทย์นี้เป็นโจทย์สัมภาษณ์งานเขียนโปรแกรมที่คลาสสิกมาก ลองทำดู!)


````python
# เขียนคำตอบที่นี่
for i in range(1,31):
  if i%3==0:
    print("Fizz")
  elif i%5==0:
    print("Buzz")
  elif i%3==0 and i%5==0:
    print("FizzBuzz")
  else:
    print(i)
````
